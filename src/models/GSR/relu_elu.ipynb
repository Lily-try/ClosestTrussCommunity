```python
root_path = '/home/jianan/Desktop/MGSL_new/'
result_folder = f'{root_path}results/GSR/cora/'
import sys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import os
sys.path.append(root_path + 'src')
os.chdir(root_path)
import seaborn as sns
from utils.proj_settings import EVAL_METRIC
import warnings
warnings.filterwarnings('ignore')
pd.set_option("display.max_rows", 10)
sns.set_theme()
```

```python
relu_file = f'{result_folder}GSR_cora<l00><ReluTune>82.80'+'.xlsx'
relu_df = pd.read_excel(relu_file).drop(columns=['Unnamed: 0'])
metric_names = [cname[4:] for cname in relu_df.columns if 'avg' in cname]
# relu_df[EVAL_METRIC] = relu_df[EVAL_METRIC].apply(
#     lambda x: float(x.split('±')[0])).astype(float)
test_acc = relu_df[EVAL_METRIC].values
acc_list, std_list = [], []
for i in range(len(test_acc)):
    acc, std = test_acc[i].split('±')
    acc_list.append(float(acc))
    std_list.append(float(std))

relu_df['accuracy'] = acc_list
relu_df['variance'] = std_list
relu_df
```

```python
elu_file = f'{result_folder}GSR_cora<l00><EluTune>82.68'+'.xlsx'
elu_df = pd.read_excel(elu_file).drop(columns=['Unnamed: 0'])
metric_names = [cname[4:] for cname in elu_df.columns if 'avg' in cname]
# elu_df[EVAL_METRIC] = elu_df[EVAL_METRIC].apply(
#     lambda x: float(x.split('±')[0])).astype(float)
test_acc = elu_df[EVAL_METRIC].values
acc_list, std_list = [], []
for i in range(len(test_acc)):
    acc, std = test_acc[i].split('±')
    acc_list.append(float(acc))
    std_list.append(float(std))

elu_df['accuracy'] = acc_list
elu_df['variance'] = std_list
elu_df
```

```python
zero_to_one_rough_list = [0.0, 0.25, 0.5, 0.75, 1.0]
fsim_list = zero_to_one_rough_list
intra_list = zero_to_one_rough_list
add_list = zero_to_one_rough_list
p_batch_list = [128, 256, 512]
p_epochs_list = [0, 10, 20, 30, 50]
```

```python
## Add Ratio
```

```python
test_df = pd.DataFrame()
intra_weight = 0.5
fsim_weight = 0.5
p_batch_size = 256
p_epochs = 20
sub_add = add_list*3

add_df_relu = relu_df.query('intra_weight == 0.25 and fsim_weight==0.25 and p_batch_size==256 and p_epochs==20', engine='python')
ori_acc_relu = add_df_relu.accuracy.values
up_acc_relu = ori_acc_relu + add_df_relu.variance.values
down_acc_relu = ori_acc_relu - add_df_relu.variance.values

add_df_elu = elu_df.query('intra_weight == 0.25 and fsim_weight==0.25 and p_batch_size==256 and p_epochs==20', engine='python')
ori_acc_elu = add_df_elu.accuracy.values
up_acc_elu = ori_acc_elu + add_df_elu.variance.values
down_acc_elu = ori_acc_elu - add_df_elu.variance.values


relu_acc = ori_acc_relu.tolist()+up_acc_relu.tolist()+down_acc_relu.tolist()
elu_acc = ori_acc_elu.tolist()+up_acc_elu.tolist()+down_acc_elu.tolist()
test_df['add_ratio'] = sub_add*2
test_df['accuracy'] = relu_acc + elu_acc
test_df['activation'] = ['relu']*len(relu_acc)+['elu']*len(elu_acc)
ax = sns.lineplot(data=test_df, x='add_ratio', y='accuracy', hue='activation')
ax.figure.set_figwidth(15.0)
ax.figure.set_figheight(8.0)
test_df
```
